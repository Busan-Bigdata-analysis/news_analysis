```python
DELTEXT = ['\n','"','\'','\xa0']
DROPTEXT = '코멘트를 추가합니다'

testText = data.newsfulltext

for idx,text in enumerate(testText):
    if text == DROPTEXT:
        data.drop(index=idx, inplace=True)
    for item in DELTEXT:
        # print(type(text),type(idx))
        text = text.replace(item,'')
    testText[idx] = text
    # break

testText
```
